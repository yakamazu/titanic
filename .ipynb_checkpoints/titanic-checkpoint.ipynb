```python
# read data
train = pd.read_csv('../train.csv')
train.head(100)
```

```python
#create dummy
df_sex_dummy = pd.get_dudf_sex_dummy = pd.get_dummies(train['Sex'], prefix='Sex', drop_first=True)
df_pclass_dummy = pd.get_dummies(train['Pclass' ], prefix='Pclass', drop_first=True)
df_sibsp_dummy = pd.get_dummies(train['SibSp' ], prefix='SibSp', drop_first=True)
df_parch_dummy = pd.get_dummies(train['Parch' ], prefix='Parch', drop_first=True)
```

```python
train_dummy_add = train.join(df_sex_dummy)
train_dummy_add = train_dummy_add.join(df_pclass_dummy)
train_dummy_add = train_dummy_add.join(df_sibsp_dummy)
#train_dummy_add = train_dummy_add.join(df_parch_dummy)
train_dummy_add
```

```python
#create x, y for model
x = train_dummy_add.drop(["PassengerId", "Survived", "Pclass", "Name", "Sex", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"],  axis=1)
y = train_dummy_add["Survived"]
```

```python
#Logistic Regression
#create model and fit
clf = LogisticRegression()
clf.fit(x, y)
clf.score(x, y)

#RandomForest
#create model and fit
forest = RandomForestClassifier(random_state=0)
forest.fit(x, y)
```

```python
#変数名とその係数を格納するデータフレーム
coeff_df = pd.DataFrame([x.columns, clf.coef_[0]]).T
coeff_df
```

```python
#predict testdate 
test = pd.read_csv('../test.csv')
test = test.fillna(train.mean())

df_sex_dummy = pd.get_dummies(test['Sex'], prefix='Sex', drop_first=True)
df_pclass_dummy = pd.get_dummies(test['Pclass' ], prefix='Pclass', drop_first=True)
df_sibsp_dummy = pd.get_dummies(test['SibSp' ], prefix='SibSp', drop_first=True)
df_parch_dummy = pd.get_dummies(test['Parch' ], prefix='Parch', drop_first=True)

test_dummy_add = test.join(df_sex_dummy)
test_dummy_add = test_dummy_add.join(df_pclass_dummy)
test_dummy_add = test_dummy_add.join(df_sibsp_dummy)
#test_dummy_add = test_dummy_add.join(df_parch_dummy)

x_test = test_dummy_add.drop(["PassengerId", "Pclass", "Name", "Sex", "SibSp", "Parch", "Ticket", "Cabin", "Embarked"],  axis=1)

x_test_predict = pd.Series(clf.predict(x_test))

result_test = pd.DataFrame({'PassengerId':test_dummy_add['PassengerId'], 'Survived':x_test_predict})
result_test.to_csv('result.csv', index=False)
```
